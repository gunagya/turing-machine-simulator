```python
file = open("decimal_mulitplication_local.txt","w")

# Parse
for i in range(0,10):    
    file.write("P1 {}__ P1 {}__ >--\n".format(i,i))

    
# End of operand 1    
file.write("P1 *__ P2 ___ >--\n")


# Copy
for i in range(0,10):    
    file.write("P2 {}__ P2 _{}_ >>-\n".format(i,i))
    
file.write("P2 ___ A ___ <<-\n")


# Reverse
for i in range(0,10):    
    file.write("A _{}_ A _{}_ <--\n".format(i,i))
    
    
# Align operands
for i in range(0,10):  
    for j in range(0,10):    
        file.write("A {}{}_ C0 {}{}_ ---\n".format(i,j,i,j))
        
        
# Starting Multiplication
for c in range(0,10):    # Carry
    for i in range(0,10):    # Input on first tape        
        for j in range(0,10):    # Input on second tape
            file.write("C{} {}{}_ C{} {}{}{} <-<\n".format(c,i,j,int((i*j+c)/10),i,j,((i*j+c)%10)))     
            
            
# Multiplication
# Carry C0, C1, C2, C3,...C9;
for c in range(0,10):    # Carry
    for i in range(0,10):    # Input on first tape        
        for j in range(0,10):    # Input on second tape
            for k in range(0,10):    # Input on third tape
                file.write("C{} {}{}{} C{} {}{}{} <-<\n".format(c,i,j,k,int((i*j+k+c)/10),i,j,((i*j+k+c)%10)))     

                
# Reach the blank on the second tape -> halt
for i in range(0,10):
    for k in range(0,10):
        file.write("B {}_{} H {}_{} ---\n".format(i,k,i,k))  

for i in range(0,10):
    file.write("B {}__ H {}__ ---\n".format(i,i))  
        
        
#End operation of 1 unit of second operand
for c in range(0,10):    # Carry
    for j in range(0,10):    # Input on second tape
        for k in range(0,10):    # Input on third tape
            file.write("C{} _{}{} C{} 0{}{} <-<\n".format(c,j,k,int((c+k)/10),j,((c+k)%10)))
            
for c in range(1,10):    # Carry
    for j in range(0,10):    # Input on second tape
        file.write("C{} _{}_ B _{}{} ><-\n".format(c,j,j,c))   # Shift to the next digit second operand  
        
for j in range(0,10):
    file.write("C0 _{}_ B _{}_ ><-\n".format(j,j))

for c in range(0,10):    # Carry
    for k in range(0,10):    # Input on third tape
        file.write("C{} __{} C{} __{} --<\n".format(c,k,int((c+k)/10),((c+k)%10))) 
        
        
# Align for next unit of second operand
for i in range(0,10): 
    for j in range(0,10):        
        for k in range(0,10):  
            file.write("B {}{}{} B {}{}{} >->\n".format(i,j,k,i,j,k))
            
for i in range(0,10): 
    for j in range(0,10):        
        file.write("B {}{}_ B {}{}_ >->\n".format(i,j,i,j))

for j in range(0,10):        
    for k in range(0,10):  
        file.write("B _{}{} C0 _{}{} <-<\n".format(j,k,j,k))
        
file.close()
```
