```python
file = open("Multiply.txt","w")

# Parse
file.write("name: Multiplication\ninit: P1\naccept: H\n")
for i in range(0,10):    
    file.write("P1,{},_,_\nP1,{},_,_,>,-,-\n".format(i,i))

    
# End of operand 1    
file.write("P1,*,_,_\nP2,_,_,_,>,-,-\n")


# Copy
for i in range(0,10):    
    file.write("P2,{},_,_\nP2,_,{},_,>,>,-\n".format(i,i))
    
file.write("P2,_,_,_\nA,_,_,_,<,<,-\n")


# Reverse
for i in range(0,10):    
    file.write("A,_,{},_\nA,_,{},_,<,-,-\n".format(i,i))
    
    
# Align operands
for i in range(0,10):  
    for j in range(0,10):    
        file.write("A,{},{},_\nC0,{},{},_,-,-,-\n".format(i,j,i,j))
        
        
# Starting Multiplication
for c in range(0,10):    # Carry
    for i in range(0,10):    # Input on first tape        
        for j in range(0,10):    # Input on second tape
            file.write("C{},{},{},_\nC{},{},{},{},<,-,<\n".format(c,i,j,int((i*j+c)/10),i,j,((i*j+c)%10)))     
            
            
# Multiplication
# Carry C0, C1, C2, C3,...C9;
for c in range(0,10):    # Carry
    for i in range(0,10):    # Input on first tape        
        for j in range(0,10):    # Input on second tape
            for k in range(0,10):    # Input on third tape
                file.write("C{},{},{},{}\nC{},{},{},{},<,-,<\n".format(c,i,j,k,int((i*j+k+c)/10),i,j,((i*j+k+c)%10)))     

                
# Reach the blank on the second tape -> halt
for i in range(0,10):
    for k in range(0,10):
        file.write("B,{},_,{}\nH,{},_,{},-,-,-\n".format(i,k,i,k))  

for i in range(0,10):
    file.write("B,{},_,_\nH,{},_,_,-,-,-\n".format(i,i))  
        
        
#End operation of 1 unit of second operand
for c in range(0,10):    # Carry
    for j in range(0,10):    # Input on second tape
        for k in range(0,10):    # Input on third tape
            file.write("C{},_,{},{}\nC{},0,{},{},<,-,<\n".format(c,j,k,int((c+k)/10),j,((c+k)%10)))
            
for c in range(1,10):    # Carry
    for j in range(0,10):    # Input on second tape
        file.write("C{},_,{},_\nB,_,{},{},>,<,-\n".format(c,j,j,c))   # Shift to the next digit second operand  
        
for j in range(0,10):
    file.write("C0,_,{},_\nB,_,{},_,>,<,-\n".format(j,j))

for c in range(0,10):    # Carry
    for k in range(0,10):    # Input on third tape
        file.write("C{},_,_,{}\nC{},_,_,{},-,-,<\n".format(c,k,int((c+k)/10),((c+k)%10))) 
        
        
# Align for next unit of second operand
for i in range(0,10): 
    for j in range(0,10):        
        for k in range(0,10):  
            file.write("B,{},{},{}\nB,{},{},{},>,-,>\n".format(i,j,k,i,j,k))
            
for i in range(0,10): 
    for j in range(0,10):        
        file.write("B,{},{},_\nB,{},{},_,>,-,>\n".format(i,j,i,j))

for j in range(0,10):        
    for k in range(0,10):  
        file.write("B,_,{},{}\nC0,_,{},{},<,-,<\n".format(j,k,j,k))
        
file.close()
```
